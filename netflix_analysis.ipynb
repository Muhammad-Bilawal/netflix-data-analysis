## 1. Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## 2. Load Dataset

```python
df = pd.read_csv('netflix1.csv')
df.head()
```

## 3. Understanding the Data
In this step, we explore the structure of the dataset.

```python
df.info()
```

```python
df.describe()
```

## 4. Checking Missing Values

## 5. Data Analysis
### 5.1 Distribution of Content Types

```python
df['type'].value_counts()
```

### 5.2 Top Countries Producing Content

```python
df['country'].value_counts().head(10)
```

### 5.3 Most Common Genres

```python
df['listed_in'].value_counts().head(10)
```

## 6. Visualization

```python
df['type'].value_counts().plot(kind='bar')
plt.title('Distribution of Movies and TV Shows')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

```python
df['country'].value_counts().head(5).plot(kind='bar')
plt.title('Top 5 Countries')
plt.xlabel('Country')
plt.ylabel('Number of Titles')
plt.show()
```

## 7. Conclusion
This analysis shows that Netflix has more movies than TV shows. The United States appears to contribute the highest number of titles. Overall, the dataset reflects global entertainment trends and content diversity.
